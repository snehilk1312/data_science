```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("List_of_Formula_One_World_Constructors'_Champions.csv")
```

```python
data.drop(['Tyre \n','Drivers [1] \n', 'Poles \n' ],axis=1,inplace=True)
```

```python
data.drop([0,63,64],inplace=True)
```

```python
data.columns = ['season', 'constructor', 'poles', 'wins', 'Podiums', 'Fastest_lap', 'Points', 'Margin_points']
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
dff = data.groupby('constructor').head(62).reset_index(drop=True)
```

```python
dff['freq'] = dff.groupby('constructor')['constructor'].transform('count')
```

```python
dff
```

```python
dff.sort_values(by=['freq'], ascending=False, inplace=True)
```

```python
dff_unique = dff.drop_duplicates(subset=['constructor'])
```

```python
dff_unique.reset_index(drop=True, inplace=True)
```
