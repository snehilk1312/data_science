```python
s=np.array([1,2,3])
t=np.array([4,5,6])
np.dot(s,t)   #If both a and b are 1-D arrays, it is inner product of vectors (without complex conjugation).
```

```python
s=np.array([[1],[2]])
t=np.array([[3],[1]])
print(s)
print(t)
```

```python
np.dot(s,t)
```

```python
s=np.array([[1,2],[3,4]])
t=np.array([[1,0],[0,1]])
np.dot(s,t)   #If both a and b are 2-D arrays, it is matrix multiplication
```

```python
s=np.array([[1,2],[3,4]])
t=np.array([5,6])
np.dot(s,t)   #If a is an N-D array and b is a 1-D array, it is a sum product over the last axis of a and b.
```
