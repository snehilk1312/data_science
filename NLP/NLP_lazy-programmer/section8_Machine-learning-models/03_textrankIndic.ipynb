### Create own tokenizer instance - Resulting in correct 5 line summarization
##### `link1`: https://indic-nlp-library.readthedocs.io/en/latest/indicnlp.tokenize.html
##### `link2`: https://github.com/miso-belica/sumy/blob/main/docs/how-to-add-new-language.md

```python
from typing import List
from indicnlp.tokenize import indic_tokenize ,sentence_tokenize

class Tokenizer:
    @staticmethod
    def to_sentences(text: str) -> List[str]:
        return sentence_tokenize.sentence_split(text, lang='hin', delim_pat='auto')

    @staticmethod
    def to_words(sentence: str) -> List[str]:
        return indic_tokenize.trivial_tokenize_indic(sentence)
```

```python
summarizer = TextRankSummarizer()
parser = PlaintextParser.from_string(
    doc.split("\n", 1)[1],
    Tokenizer())
summary = summarizer(parser.document, sentences_count=5)
```

```python
for sentence in summary:
    print(sentence)
```

### Latent Semantic Analysis - LSA

```python
summarizer = LsaSummarizer()
summary = summarizer(parser.document, sentences_count=5)
for s in summary:
  print(str(s))
```
